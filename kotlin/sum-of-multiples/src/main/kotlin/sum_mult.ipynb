```kotlin
val factors = setOf(3, 5)
val limit = 10
```

```kotlin
val posFactors: List<Int> = factors.filter {it > 0 }
posFactors
```

```kotlin
fun multiples(i: Int): List<Int> {
    return posFactors.map {if (i % it == 0) i else 0}
}
```
